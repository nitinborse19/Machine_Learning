```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("C:\\Users\\Krishna\\Desktop\\dataset\\Bodyfat_1.csv")
df
```

## Pair Plot all Dataset

```python
#Plot pair plot on data set
sns.set_style(style='darkgrid')
sns.pairplot(df)
```

#### Assigning All Columns variable name

```python
x=df['bodyfat']
y=df['Age']
z=df['Weight']
A=df['Height']
B=df['Density']
```

## Scatter Plot 

```python
#scatter plot on data set
plt.scatter(x,y,marker='o',color='b',label='bodyfat')
plt.scatter(z,y,marker='o',color='y',label='Weight')
plt.scatter(A,y,marker='o',color='g',label='Height')
plt.scatter(B,y,marker='o',color='r',label='Density')
plt.legend()
plt.xlabel('Density,bodyfat, Age ,Weight ,Height')
plt.ylabel('Age')
plt.show
```
