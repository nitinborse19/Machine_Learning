```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["R"]=le.fit_transform(df["R"])
```

```python
df
```

```python
x=df.drop(columns=["R"])
y=df["R"]
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.svm import SVC
pipeline_SVC=Pipeline([('scaler1',StandardScaler()),
                     ('pca1',PCA(n_components=2)),
                     ('GNB',SVC(kernel="rbf",C=10,gamma="scale",degree=4))])
```

```python
from sklearn.neighbors import KNeighborsClassifier
pipeline_KNN=Pipeline([('scaler1',StandardScaler()),
                     ('pca1',PCA(n_components=2)),
                     ('GNB',KNeighborsClassifier(n_neighbors=6,weights="distance"))])
```

```python
pipeline_GN=Pipeline([('scaler1',StandardScaler()),
                     ('pca1',PCA(n_components=2)),
                     ('GNB',GaussianNB())])
```

```python
pipeline_Dt=Pipeline([('scaler2',StandardScaler()),
                     ('pca2',PCA(n_components=2)),
                     ('Dt_classifier',DecisionTreeClassifier(criterion="entropy",splitter="best",max_depth=5))])
```

```python
pipeline_randomforest=Pipeline([('scaler3',StandardScaler()),
                     ('pca3',PCA(n_components=2)),
                     ('rf_classifier',RandomForestClassifier(n_estimators=100,max_depth=3,max_features="sqrt"))])
```

```python
## Lets make the list of Pipeline
pipelines=[pipeline_SVC,pipeline_KNN,pipeline_GN,pipeline_Dt,pipeline_randomforest]
```

```python
best_accuracy=0.0
best_Regression=0
best_pipeline=""
```

```python
# Dictionary of pipeline and classifier types for ease of reference
pipe_dict={0:"GNB",1:"Decision Tree",2:"Random Forest",3:"SVC",4:"KNN"}

# fit the pipelines
for pipe in pipelines:
    pipe.fit(xtrain,ytrain)
```

```python
for i,model in enumerate(pipelines):
    print("{} Test accuracy: {}".format(pipe_dict[i],model.score(xtest,ytest)))
```

```python
for i,model in enumerate(pipelines):
    if model.score(xtest,ytest)>best_accuracy:
        best_accuracy=model.score(xtest,ytest)
        best_pipeline=model
        best_classifier=i
print('Classifier with best accuracy:{}'.format(pipe_dict[best_classifier]))
```
