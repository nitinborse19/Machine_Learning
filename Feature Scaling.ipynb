```python
# Feature scaling
#Rescaling (min-max normalization)
#[0,1]=x-xmin/xmax-xmin
#[a,b]=a+(x-xmin)*(b-a)/Xmax-Xmin
#standardization(Z-scale normalization)
```

```python
# rescaling(min-max normalization)
import numpy as np
data=np.array([1.4,3.4,5.6,2.3])
print(data)
```

```python
dmin=np.min(data)
dmax=np.max(data)
```

```python
# X-Xmin/Xmax-Xmin                            its possible bcoz of numpy array
print((data-dmin)/(dmax-dmin))
```

```python
data=data.reshape(-1,1)            # -1 is row and 1 is column
from sklearn.preprocessing import MinMaxScaler
m=MinMaxScaler()
m.fit(data)
print(m.transform(data))
```

# Standardization

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
data=np.array([1.3,2.1,4.5,6.5])
print("Before scaling")
print(data)
mean=np.mean(data)
std=np.std(data)
print(np.mean(data))
print(np.std(data))
#data-mean/std
scaled_data=(data-mean)/std
print("after scaling")
print(scaled_data)
print(np.mean(scaled_data))
print(np.std(scaled_data))
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
data=np.array([1.3,2.1,4.5,6.5])
data=data.reshape(-1,1)            # -1 is row and 1 is column
m=StandardScaler()
m.fit(data)
sc_data=m.transform(data)
print(np.mean(sc_data))
print(np.std(sc_data))
#fit_transform
sc_data1=m.fit_transform(data)
print(np.mean(sc_data1))
print(np.std(sc_data1))
```
