# Iris Dataset 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Analys Dataset

```python
df=sns.load_dataset("iris")
df
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['label_species']=le.fit_transform(df['species'])
```

```python
df.head()
```

## 1.Build a model with raw data

```python
x=df.drop(columns=["species","label_species"])
y=df["label_species"]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier(n_neighbors=4)
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
ypred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
acc=accuracy_score(ytest,ypred)
acc
```

```python
cm=confusion_matrix(ytest,ypred)
cm
```

```python
report=classification_report(ytest,ypred)
print(report)
```

## 3.Check the performance on Test dataset

```python
#neibors
#metrics
# weigth  2 types 1) uniform 2)distance
```

```python
# estamitor
model=KNeighborsClassifier()
# parameter
n_neighbors=range(1,31)
metric=['eucludean','manhattan','minkowski']
weights=['uniform','distance']
#parameter grid
grid={"n_neighbors":n_neighbors,"metric":metric,"weights":weights}

# cross Validator
from sklearn.model_selection import RepeatedStratifiedKFold
kfold_cv=RepeatedStratifiedKFold(n_splits=10,n_repeats=5,random_state=0)

#grid_search_cv
from sklearn.model_selection import GridSearchCV
grid_cv=GridSearchCV(estimator=model,param_grid=grid,cv=kfold_cv,scoring="accuracy",n_jobs=1)
res=grid_cv.fit(xtrain,ytrain)
```
