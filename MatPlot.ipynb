# Data Visualization 

```python
import matplotlib.pyplot as plt
x=[1,3,5,6]         
y=[4,5,3,2]

#lineplot ==>
plt.plot(x,y)
plt
```

#  o,*,^,+ ---> shapes     or m,b,g,y,k --> color

```python
plt.plot(x,y,marker="o")
plt
```

```python
plt.plot(x,y,marker="+",color="r")
plt
```

```python
plt.plot(x,y,marker="o",color="g",label="myline")          #its used for label plt.legend()
#plt.legend()
plt
```

```python
plt.plot(x,y,marker="o",color="g",label="myline")
plt.plot(y,x,marker="o",color="b",label="myline")
plt.legend()
plt
```

```python
plt.plot(x,y,marker="o",color="g",label="myline")
plt.legend()
plt
```

```python
plt.plot(y,x,marker="+",color="b",label="myline")
plt.legend()
plt
```

```python
plt.plot(x,y,marker="o",color="r",label="myline")
plt.plot(y,x,marker="o",color="b",label="myline",linestyle="--")
plt.legend()
plt.title("my first line plot")
plt.xlabel("my x axis")
plt.ylabel("my y axis")
plt
```

```python
plt.plot(x,y,marker="o",color="b",label="myline",linestyle="--",markerfacecolor="y")
plt.legend()
plt.title("my first linr plot")
plt.xlabel("my x axis")
plt.ylabel("my y axis")
plt.xticks([2,4,6])
plt
plt
```

```python
plt.plot(x,y,marker="o",color="b",label="myline",linestyle="--",markerfacecolor="y")
plt.legend()
plt.title("my first linr plot")
plt.xlabel("my x axis")
plt.ylabel("my y axis")
plt.xticks([2,4,6])
plt.yticks([2,4,6])
plt
```

```python
plt.plot(x,y,marker="o",color="b",label="myline",linestyle="--",markerfacecolor="y")
plt.legend(loc="best")
plt.title("my first linr plot")
plt.xlabel("my x axis")
plt.ylabel("my y axis")
plt.xticks([2,4,6])
plt.yticks([2,4,6])
plt
```

```python
plt.plot(x,y,marker="o",color="b",label="myline",linestyle="--",markerfacecolor="y")
plt.legend(loc="lower center")
plt.title("my first line plot")
plt.xlabel("my x axis")
plt.ylabel("my y axis")
plt.xticks([2,4,6],["a","b","c"])
plt.yticks([2,4,6])
plt
```
