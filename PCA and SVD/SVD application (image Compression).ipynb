```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("../Image/lotus-flower.jpg")
```

```python
type(img)
```

```python
img.shape
```

```python
#convert the image to gray scale
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

plt.imshow(gray,cmap="gray")
```

```python
U,S,V =np.linalg.svd(gray,full_matrices=False)
print(U.shape,S.shape,V.shape)
```

```python
comps=[U.shape[0],100,50,20,10,5,2,1]
fig=plt.figure(figsize=(15,7))
for i in range(len(comps)):
    trans_image=U[:,:comps[i]]@ np.diag(S[:comps[i]]) @ V[:comps[i],:]
    plt.subplot(2,4,i+1)
    plt.imshow(trans_image,cmap="gray")
    plt.title("image with n_components "+str(comps[i]))
```
