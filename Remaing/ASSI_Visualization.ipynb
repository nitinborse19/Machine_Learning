# Data Visualization

```python
#Plot pair plot on data set
sns.set_style(style='darkgrid')
sns.pairplot(krish)
```

```python
krish.columns
```

```python
x=krish['bodyfat']
y=krish['Age']
z=krish['Weight']
A=krish['Height']
B=krish['Density']
```

```python
#scatter plot on data set
plt.scatter(x,y,marker='o',color='b',label='bodyfat')
plt.scatter(z,y,marker='o',color='y',label='Weight')
plt.scatter(A,y,marker='o',color='g',label='Height')
plt.scatter(B,y,marker='o',color='r',label='Density')
plt.legend()
plt.xlabel('Density,bodyfat, Age ,Weight ,Height')
plt.ylabel('Age')
plt.show
```

```python
print(krish['bodyfat'].sum())
print(krish['Age'].sum())
print(krish['Weight'].sum())
print(krish['Height'].sum())
```

```python
# pie chart
%matplotlib inline
P=[4826,11311,45088.9,17677.5]
Q=['bodyfat', 'Age' ,'Weight' ,'Height']
plt.axis('equal')
plt.pie(P,labels=Q,radius=2 ,autopct='%0.1f%%',shadow=True,explode=[0.5,0,0,0])
plt.show()
```
