```python
df["State/UnionTerritory"].unique()
```

```python
df=df.replace('Telengana','Telangana')
df=df.replace('Telengana***','Telangana')
df=df.replace('Maharashtra***','Maharashtra')
df=df.replace('Chandigarh***','Chandigarh')
df=df.replace('Punjab***','Punjab')
```

```python
df["State/UnionTerritory"].unique()
```

```python
df.describe()
```

### Calculate Total Confirmed, Cured , Deaths and Active Cases

```python
x=df["Confirmed"].sum()
y=df["Cured"].sum()
z=df["Deaths"].sum()
active=x-y
print("Total Confirmed cases=",x)
print("Total Cured cases=",y)
print("Total Active cases=",active)
print("Total Deaths cases=",z)
```

```python
sns.barplot(x=['Confirmed','Cured','Deaths','active'],y=[x,y,z,active])
```

## Top 10 Contaminated State

###  Most affected state --->  Maharashtra State Covid Dataset

```python
mh_df=df[df["State/UnionTerritory"]=="Maharashtra"]
mh_df
```

### Below we have done data filtering as we want a time series data so we have decided a date and filtered that data only...

```python
ret_df=mh_df[mh_df["Date"].isin(['09/03/20','09/04/20','09/05/20','09/06/20','09/07/20','09/08/20','09/09/20','09/10/20','09/11/20','09/12/20'])]
```

```python
ret_df
```

### Data Visualization Confirmed cases Vs Cured Cases Vs Deaths Maharashtra State

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=ret_df,x='Date',y='Confirmed',color="b",label="Confirmed Cases")
sns.lineplot(data=ret_df,x='Date',y='Cured',color="g",label="Cured Cases")
sns.lineplot(data=ret_df,x='Date',y='Deaths',color="r",label="Deaths Cases")
plt
```

## Now we will line plot time series of all states data

```python
ret_df2=df[df["Date"].isin(['09/03/20','09/04/20','09/05/20','09/06/20','09/07/20','09/08/20','09/09/20','09/10/20','09/11/20','09/12/20'])]
ret_df2
```

### Date wise Confirmed Cases all State dataset
