# Plot scatter plot and pair plot on BodyFat Dataset

```python
#Bodyfat Dataset Assignment

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("E:\\Dataset\\Bodyfat.csv")
df
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
#Pair plot All Dataset
sns.set_style(style="darkgrid")
sns.pairplot(df)
```

```python
#Assigning All Columns variable name
x=df["bodyfat"]
y=df["Age"]
z=df["Weight"]
A=df["Height"]
B=df["Density"]
```

```python
#Scatter plot
plt.scatter(x,y,marker="*",color="g",label="bodyfat")
plt.scatter(z,y,marker="*",color="y",label="Weight")
plt.scatter(A,y,marker="*",color="r",label="Height")
plt.scatter(B,y,marker="*",color="b",label="Density")
plt.legend()
plt.xlabel("Density,bodyfat,Age,Weight,Height")
plt.ylabel("Age")
plt.show()
```

```python
a=df["bodyfat"].sum()
print(a)
b=df["Age"].sum()
print(b)
c=df["Weight"].sum()
print(c)
d=df["Height"].sum()
print(d)
```

```python
#Bar Plot
sns.barplot(x=["bodyfat","Age","Weight","Height"],y=[a,b,c,d])
```

#  Use covid data set to visualize overall and state wise data and generate report.


```python
df=pd.read_csv("E://Dataset//covid_19_india.csv")
df
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
df.head()
```

```python
#Confirmed Cases

plt.figure(figsize=(10,5))
plt.plot(df["Confirmed"])
plt.title("confirmed Cases")
plt.show()
```

```python
#Deaths Cases

plt.figure(figsize=(10,5))
plt.plot(df["Deaths"])
plt.title("Deaths Cases")
plt.show()
```

```python
df["State/UnionTerritory"].unique()
```

```python
df=df.replace("Telengana","Telangana")
df=df.replace("Telengana***","Telangana")
df=df.replace("Maharashtra***","Maharashtra")
df=df.replace("Chandigarh***","Chandigarh")
df=df.replace("Punjab***","Punjab")
```

```python
df["State/UnionTerritory"].unique()
```

```python
df.describe()
```

```python
#Calculate Total Confirmed,Cured,Deaths and Active Cases
x=df["Confirmed"].sum()
y=df["Cured"].sum()
z=df["Deaths"].sum()
active=x-y
print("Total Confirmed cases=",x)
print("Total Cured cases=",y)
print("Total Active cases=",active)
print("Total Deaths cases=",z)
```

```python
sns.barplot(x=['Confirmed','Cured','Deaths','active'],y=[x,y,z,active])
```

```python
df1=df.tail(35)
df2=df1.sort_values(by="Confirmed",ascending=False).head(10)
print(df2)
```

```python
#Highest Confirmed Cases Top 10 State
plt.figure(figsize=(20,10))
plt.bar(df2["State/UnionTerritory"],height=df2["Confirmed"],color="y")
```

```python
#Most affected state==> Maharashtra State Covid Dataset
mh_df=df[df["State/UnionTerritory"]=="Maharashtra"]
print(mh_df)
```

```python
#Below we have done data filtering as we want a time series data so we have decided a date and filtered that data only...
rt_df=mh_df[mh_df["Date"].isin(['09/03/20','09/04/20','09/05/20','09/06/20','09/07/20','09/08/20','09/09/20','09/10/20','09/11/20','09/12/20'])]
```

```python
print(rt_df)
```

```python
#Data Visualization Confirmed cases Vs Cured Cases Vs Deaths Maharashtra State

plt.figure(figsize=(10,5))
sns.lineplot(data=rt_df,x='Date',y='Confirmed',color="b",label="Confirmed Cases")
sns.lineplot(data=rt_df,x='Date',y='Cured',color="g",label="Cured Cases")
sns.lineplot(data=rt_df,x='Date',y='Deaths',color="r",label="Deaths Cases")
plt.show()
```

```python
#Now we will line plot time series of all states data

rt_df2=df[df["Date"].isin(['09/03/20','09/04/20','09/05/20','09/06/20','09/07/20','09/08/20','09/09/20','09/10/20','09/11/20','09/12/20'])]
print(rt_df2)
```

```python
#Date wise Confirmed Cases all State dataset

plt.figure(figsize=(20,10))
sns.pointplot(data=rt_df2,x="Date",y="Confirmed",hue="State/UnionTerritory")
plt.legend(bbox_to_anchor=(1,1))
plt.title("Confirmed Cases")
plt.show()
```

```python
#Date wise Deaths Cases all State dataset

plt.figure(figsize=(20,10))
sns.pointplot(data=rt_df2,x="Date",y="Deaths",hue="State/UnionTerritory")
plt.legend(bbox_to_anchor=(1,1))
plt.title("Deaths Cases")
plt.show()
```

```python
#Date wise Cured Cases all State dataset

plt.figure(figsize=(20,10))
sns.pointplot(data=rt_df2,x="Date",y="Cured",hue="State/UnionTerritory")
plt.legend(bbox_to_anchor=(1,1))
plt.title("Cured Cases")
plt.show()
```
