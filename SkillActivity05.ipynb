# Seasonality

```python
df_month = df.resample("M").mean()
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(df_month['2017':].index, df_month.loc['2017':, "Vancouver"], width=25, align='center')
```

```python
df_month['Vancouver'].plot(figsize=(8, 6))
```

```python
week = df.resample("W").mean()
```

```python
week
```

```python
fig=plt.figure(figsize=(20,10))
plt.plot(df.loc['2016-01':'2016-08', 'Vancouver'], marker='.', linestyle='-', linewidth = 0.5, label='Daily', color='black')
plt.plot(week.loc['2016-01':'2016-08', 'Vancouver'], marker='o', markersize=8, linestyle='-', label='Weekly', color='coral')
plt.legend()
```

```python
roll_10=df.rolling(7, center=True).mean()
```

```python
fig=plt.figure(figsize=(20,10))
plt.plot(df.loc['2016-01':'2016-08', 'Vancouver'], marker='.', linestyle='-', linewidth = 0.5, label='Daily', color='black')
plt.plot(week.loc['2016-01':'2016-08', 'Vancouver'], marker='o', markersize=8, linestyle='-', label='Weekly', color='coral')
plt.plot(roll_10.loc['2016-01':'2016-08', 'Vancouver'], marker='o', markersize=8, linestyle='-', label='Weekly', color='red')
plt.ylabel("Open")
plt.legend()
```

# Decomposition

```python
from pylab import rcParams
import statsmodels.api as sm
rcParams['figure.figsize'] = 11, 9
decomposition = sm.tsa.seasonal_decompose(df_month['Vancouver'], model='Additive')
fig = decomposition.plot()
```

#  Check Data is stationary or Not

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(df["Vancouver"])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
	print('\t%s: %.3f' % (key, value))
```
