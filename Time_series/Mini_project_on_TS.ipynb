## Using Moving Average 

```python
df=pd.read_csv("C:\\Users\\Krishna\\Desktop\\dataset\\TCS.NS.csv",parse_dates=['Date'])
df
```

```python
df=df.dropna()
```

```python
df.info()
```

### Simple Moving Average Technique

```python
df["SMA2"]=df["Volume"].rolling(window=2).mean()
```

```python
df
```

```python
df["SMA4"]=df["Volume"].rolling(window=4).mean()
```

```python
df
```

```python
df["SMA7"]=df["Volume"].rolling(window=7).mean()
```

```python
df
```

```python
p_df=df.head(100)
p_df
```

```python
fig=plt.figure(figsize=(10,5))
plt.plot(p_df["Volume"],label="Original")
plt.plot(p_df["SMA2"],label="SMA2")
plt.plot(p_df["SMA4"],label="SMA4")
plt.plot(p_df["SMA7"],label="SMA7")
plt.legend()
```

## Commulative Moving Average Technique

```python
df["CMA"]=df["Volume"].expanding(min_periods=5).mean()
```

```python
df=df.head(100)
```

```python
df["CMA3"]=df["Volume"].expanding(min_periods=3).mean()
```

```python
fig=plt.figure(figsize=(10,5))
plt.plot(df["Volume"],label="Original")
plt.plot(df["CMA"],label="CMA")
plt.plot(df["CMA3"],label="CMA3")
plt.legend()
```

```python
#Exponential Moving Avarege

# Exponential Weighted a function
# 2 adjust-- True 
# Com,span,halflife--alpha
# directly can use alpha if its given or you can do a trail error to see the smoothing effect
```

```python
df["EMA"]=df["Volume"].ewm(com=0.5,adjust=True).mean()
```

```python
df.head()
```

```python
fig=plt.figure(figsize=(10,5))
plt.plot(df["Volume"],label="Original")
plt.plot(df["EMA"],label="EMA")
plt.legend()
```

```python
df["EMA1"]=df["Volume"].ewm(com=0.2,adjust=True).mean()
```

```python
df_b=df.head(100)
```

```python
fig=plt.figure(figsize=(10,5))
plt.plot(df_b["Volume"],label="Original")
plt.plot(df_b["CMA"],label="CMA")
plt.plot(df_b["EMA1"],label="EMA1")
plt.legend()
```

## Conclusion
