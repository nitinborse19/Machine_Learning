```python
# Commulative 
```

```python
df["CMA"]=df["Boston"].expanding(min_periods=4).mean()
```

```python
df.head()
```

```python
df["CMA3"]=df["Boston"].expanding(min_periods=3).mean()
```

```python
fig=plt.figure(figsize=(10,5))
plt.plot(df["Boston"],label="Original")
plt.plot(df["CMA"],label="CMA")
plt.plot(df["CMA3"],label="CMA3")
plt.legend()
```

```python
#Exponential Moving Avarege

# Exponential Weighted a function
# 2 adjust-- True 
# Com,span,halflife--alpha
# directly can use alpha if its given or you can do a trail error to see the smoothing effect
```

```python
df["EMA"]=df["Boston"].ewm(com=0.5,adjust=True).mean()
```

```python
df.head()
```

```python
fig=plt.figure(figsize=(10,5))
plt.plot(df["Boston"],label="Original")
plt.plot(df["EMA"],label="EMA")
plt.legend()
```

```python
df["EMA1"]=df["Boston"].ewm(span=20).mean()
```

```python
df_b=df.head(100)
```

```python
fig=plt.figure(figsize=(10,5))
plt.plot(df_b["Boston"],label="Original")
plt.plot(df_b["CMA"],label="CMA")
plt.plot(df_b["EMA1"],label="EMA1")
plt.legend()
```

```python
df["EMA2"]=df["Boston"].ewm(alpha=0.6).mean()
```

```python
df_c=df.head(100)
```

```python
fig=plt.figure(figsize=(10,5))
plt.plot(df_c["Boston"],label="Original")
plt.plot(df_c["CMA"],label="CMA")
plt.plot(df_c["EMA2"],label="EMA2")
plt.legend()
```
